### No.1: Jeeho's destination data set

```python
# import pandas
import pandas as pd
# load the data set: The data set has 5 rows and 7 columns
Dest = pd.read_csv("D:/A-study/analytics programming/Jeeho.csv",index_col= 0)
Dest
```

#### 1. Data tidy and transform

```python
# apply pd.melt function to put the last 6 destination variables into single one column
Dest = Dest.melt(id_vars ='User Id', value_vars = ['Sports','Religious','Nature','Theatre','Shopping','Picnic'])
Dest
```

```python
# rename the 3rd column from variable to Destination
Dest.rename({'variable':'Destination'}, axis='columns', inplace=True)
```

```python
# transform long format to wide
# group Dest by 'User Id' and 'Destination', calculate sum value and unstack the format to the original.
Dest_origin = Dest.groupby(['User Id', 'Destination'])['value'].sum().unstack()
Dest_origin
```

#### 2. Perform data analysis

The request: which class among the six has the highest frequency of reviews by the users.

```python
Dest.groupby('Destination')['value'].sum().sort_values(ascending=False)
```

#### Result: From the chart above, theatre has the highest frequency of reviews by the users.

### No.2: Natan's math grades dataset

```python
# load the data set: The data set has 8 rows and 5 columns
math = pd.read_csv("D:/A-study/analytics programming/natan.csv")
math
```

#### 1. Data tidy

```python
# apply pd.melt function to put the last 6 destination variables into single one column
math = math.melt(id_vars ='Name', value_vars = ['Test 1','Test 2','Test 3','Test 4'])
math
```

```python
# rename the 3rd column from variable to Test Name.
math.rename({'variable':'Test Name'}, axis='columns', inplace=True)
```

```python
math
```

```python
# transform long format to wide
# group math data by 'Name' and 'Test Name', and unstack the format to the original.
math_origin = math.groupby(['Name', 'Test Name'])['value'].mean().unstack()
math_origin
```

#### 2. Perform data analysis

Q1. Which student had the highest average test grades?

```python
avg = math.groupby('Name')['value'].mean()
avg.sort_values(ascending=False)
```

Result: Jim Jackson had the highest average test grades.

Q2. Which student had the greatest range of test grades?

```python
grade_range = math.groupby('Name')['value'].max() - math.groupby('Name')['value'].min()
grade_range.sort_values(ascending=False)
```

Result: Reggie Miller and Chris Webber shared the greatest range of test grades.

Q3. Which test had the highest grades?

```python
highest = math.groupby('Test Name')['value'].max()
highest.sort_values(ascending=False)
```

Result: Test 1 had the highest grades.

### No.3: Qi's student enrollment dataset

```python
# load the data set: The data set has 7 rows and 11 columns
enrol = pd.read_csv("D:/A-study/analytics programming/qi.csv")
enrol
```

#### 1. Data tidy and transform

```python
# apply pd.melt function to put the last 9 destination variables into single one column
enrol = enrol.melt(id_vars =['Unique ID of the institution','Level of student'])
enrol
```

```python
# rename the 4th column from variable to Test Name.
enrol.rename({'variable':'race/ethnicity'}, axis='columns', inplace=True)
```

```python
enrol.head()
```

```python
# transform long format to wide
# group enrollment data by 'Unique ID of the institution', 'Level of student', and'race/ethnicity', then unstack the format to the original.
enrol_origin = enrol.groupby(['Unique ID of the institution', 'Level of student','race/ethnicity'])['value'].mean().unstack()
enrol_origin
```

#### 2. Perform data analysis

Q1: Analyze the enrollment on different ethnicity by the level

```python
# group enrollment data by 'Level of student' and 'race/ethnicity', then unstack the format to the original.
enrol_by_level_and_race = enrol.groupby(['Level of student','race/ethnicity'])['value'].agg(sum).unstack('Level of student')
enrol_by_level_and_race.sort_values(by=['Graduate','Undergraduate'], ascending=False)
```

Result: The white race has the highset enrollment number in both graduate level and undergraduate level.

Q2: Compare different institutions’ enrollment by ethnicity

```python
# group enrollment data by 'institution' and 'race/ethnicity', then unstack the format to the original.
enrol_by_instit_and_race = enrol.groupby(['Unique ID of the institution','race/ethnicity'])['value'].agg(sum).unstack()
enrol_by_instit_and_race
```

The above chart shows the comparison of enrollment numbers of different race/ethnicity for several institutions.
